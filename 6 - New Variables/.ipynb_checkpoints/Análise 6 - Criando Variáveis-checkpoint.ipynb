```python
import pandas as pd
```

```python
dados = pd.read_csv('Data/aluguel_res_sem_missing.csv', sep =';')
```

```python
dados.head()
```

```python
# Variável Valor Bruto

dados['Valor Bruto'] = (dados['Valor'] + dados['Condominio'] + dados['IPTU']).round(2)

# Variável Metro Quadrado

dados['Valor Metro Quadrado'] = (dados['Valor'] / dados['Area']).round(2)

# Variável Valor Bruto M2

dados['VB por Metro Quadra'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
casa = ['Casa', 'Casa de Condominio', 'Casa de Vila']

dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento' )
```

```python
dados
```

# Excluindo Variáveis

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor Bruto', 'VB por Metro Quadra', 'Tipo Agregado']])
```

```python
dados_aux
```

```python
del dados_aux['Valor Bruto']

dados_aux.pop('VB por Metro Quadra')
```

```python
dados.drop(['Valor Bruto','VB por Metro Quadra'], axis =1, inplace = True)
```

```python
dados
```

```python
dados.to_csv('Data/clean_dataset.csv', sep = ';', index = False)
```
